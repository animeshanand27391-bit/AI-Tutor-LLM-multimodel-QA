# Multi-Model Q&A
Ask one question. Get answers from **Claude**, **Gemini**, and **Llama** — side by side.

```python
import os
import sys
sys.path.insert(0, '../src')

import anthropic
import google.generativeai as genai
import ollama as ollama_client
from dotenv import load_dotenv

load_dotenv('../.env')
```

```python
MODEL_ANTHROPIC = 'claude-sonnet-4-6'
MODEL_GOOGLE    = 'models/gemini-2.0-flash-lite'
MODEL_LLAMA     = 'llama3.2'

QUESTION = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

print('Question:', QUESTION)
```

```python
# --- Anthropic Claude ---
client = anthropic.Anthropic(api_key=os.environ['ANTHROPIC_API_KEY'])
response = client.messages.create(
    model=MODEL_ANTHROPIC,
    max_tokens=1024,
    messages=[{'role': 'user', 'content': QUESTION}],
)
claude_answer = response.content[0].text

print('-' * 60)
print('  Anthropic Claude')
print('-' * 60)
print(claude_answer)
```

```python
# --- Google Gemini ---
genai.configure(api_key=os.environ['GOOGLE_API_KEY'])
model = genai.GenerativeModel(MODEL_GOOGLE)
response = model.generate_content(QUESTION)
gemini_answer = response.text

print('-' * 60)
print('  Google Gemini')
print('-' * 60)
print(gemini_answer)
```

```python
# --- Ollama / Llama 3.2 (local) ---
response = ollama_client.chat(
    model=MODEL_LLAMA,
    messages=[{'role': 'user', 'content': QUESTION}],
)
llama_answer = response['message']['content']

print('-' * 60)
print('  Ollama (llama3.2)')
print('-' * 60)
print(llama_answer)
```
